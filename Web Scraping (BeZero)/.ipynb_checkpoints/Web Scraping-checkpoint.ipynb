```python
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.webdriver import WebDriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
#from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
import re
import json
```

```python
#using Selenium to scrape hidden data
url = 'https://bezerocarbon.com/listings/'
options = Options()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--ignore-ssl-errors')

# Use Selenium to open the webpage
#driver_path = r'C:\Program Files\Google\Chrome\Application\Chrome.exe'  # Replace with the actual file path
#driver = webdriver.Chrome(driver_path, options=options)
driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()

```

```python
# Fill the form fields with the required data
first_name_field = driver.find_element(By.CSS_SELECTOR,'input[name="firstname"]')#find form with CSS Selector
first_name_field.send_keys('John')#enter 'John' into thge form

last_name_field = driver.find_element(By.CSS_SELECTOR,'input[name="lastname"]')
last_name_field.send_keys('Doe')

company_field = driver.find_element(By.CSS_SELECTOR,'input[name="company"]')
company_field.send_keys('ABC Company')

email_field = driver.find_element(By.CSS_SELECTOR,'input[name="email"]')
email_field.send_keys('john.doe@example.com')

job_title_field = driver.find_element(By.CSS_SELECTOR,'input[name="jobtitle"]')
job_title_field.send_keys('Software Engineer')

industry_type_field = driver.find_element(By.CSS_SELECTOR,'select[name="industry_type"]')
industry_type_field.send_keys('Developer')

terms_checkbox = driver.find_element(By.CSS_SELECTOR,'input[name="terms_of_use_and_disclaimer"]')
driver.execute_script("arguments[0].click();", terms_checkbox)  # Use JavaScript click to select the checkbox

consent_checkbox = driver.find_element(By.CSS_SELECTOR,'input[name="legal_consent"]')
driver.execute_script("arguments[0].click();", consent_checkbox)  # Use JavaScript click to select the checkbox

# Submit the form
submit_button = driver.find_element(By.XPATH,'//button[contains(text(), "Submit")]')
driver.execute_script("arguments[0].click();", submit_button)  # Use JavaScript click to submit the form
```

```python
#CURRENTOY WORKING
result = requests.get(url) #make a request to website to receive response
soup = BeautifulSoup(result.text,'lxml') #lxml parser to get the soup (object containing all data in nested structure)
table = soup.find('table', role='table') #find all data under table section
headers = table.find_all('th') #find all contents with <th> tag in table

#Creating column names, and dataframe
titles = [i.text for i in headers]

df = pd.DataFrame(columns=titles)
df.rename({'':'Rating Watch'}, axis=1, inplace=True)

#Using Selenium to click on expand buttons, scrape hidden contents, click next page
#Using Beautifulsoup to scrape all visible contents
descriptionlist = []
buttonclk = 0 #counts number of button clicks
nextclk = 0 #counts number of button clicks
while True:
    for i in range(1, 11):
        try:
            # Button location
            button = driver.find_element(By.CSS_SELECTOR, f'#__next > main > section:nth-child(3) > div.Row__RowStyled-sc-dw8jw7-0.DOkyP.Listing__StyledRowDesktop-sc-1n7toyb-0.bjiAzY > div > table > tbody > tr:nth-child({i}) > td.name-td > div')

            # Scroll to the button
            driver.execute_script("arguments[0].scrollIntoView();", button)
            time.sleep(0.75)
            driver.execute_script("window.scrollBy(0, -300)", "")
            time.sleep(0.75)

            # Click on the button
            button.click()
            buttonclk +=1

            # Locate description and scrape hidden content
            Description = driver.find_element(By.CSS_SELECTOR, f'#__next > main > section:nth-child(3) > div.Row__RowStyled-sc-dw8jw7-0.DOkyP.Listing__StyledRowDesktop-sc-1n7toyb-0.bjiAzY > div > table > tbody > tr:nth-child({i}) > td.name-td > p')
            descriptionlist.append(Description.text)

        except NoSuchElementException:
            print("Button not found. Stopping the loop. Click count: "+str(buttonclk))
            break
                
    # Collect page elements to scrape all visible contents
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')  # lxml parser to get the soup (object containing all data in nested structure)
    table = soup.find('table', role='table')
    rows = table.find_all('tr')  # find row data with tr tags

    for row in rows[1:]:  # skips the first row which is the header
        tdata = row.find_all('td')
        row_data = [tr.text for tr in tdata]
        l = len(df)
        df.loc[l] = row_data  # update database        
            

    # Done scraping. Moving to next page.
    try:
        nextb = driver.find_elements(By.CSS_SELECTOR, '#__next > main > section:nth-child(3) > div.Row__RowStyled-sc-dw8jw7-0.DOkyP.Listing__StyledRowDesktop-sc-1n7toyb-0.bjiAzY > div > div.Row__RowStyled-sc-dw8jw7-0.cOVijK > div > button:nth-child(5)')

        if not nextb:
            print("Reached the last page. Pages count" +str(nextclk))
            break

        # Scroll to the button
        driver.execute_script("arguments[0].scrollIntoView();", nextb[0])
        time.sleep(0.75)
        driver.execute_script("window.scrollBy(0, -300)", "")
        time.sleep(0.75)

        # Click the button
        nextb[0].click()
        nextclk +=1

    except NoSuchElementException:
        print("Button not found. Stopping the loop.")
        break
```
